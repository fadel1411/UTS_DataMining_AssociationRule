```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
path = "/content/drive/MyDrive/python/dataset_0161.csv"
df = pd.read_csv(path)
df
```

```python
df.drop('date_time', inplace=True, axis=1)
df.drop('period_day', inplace=True, axis=1)
df.drop('weekday_weekend', inplace=True, axis=1)
df.head()
```

```python
onehot = df.pivot_table(index='Transaction', columns='Item', aggfunc=len, fill_value=0)
onehot.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket = onehot.applymap(encode_units)
basket.head()
```

```python
frequent_itemsets = apriori(basket, min_support=0.04, max_len=5, use_colnames= True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, min_threshold=0.1)
rules
```

```python
rules[ (rules['support'] >= 0.05) & (rules['confidence'] >= 0.1) ]
```
